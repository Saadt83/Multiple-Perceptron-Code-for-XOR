# Step_Function

```python
learning_rate=0.2
threshold=0
def step(num):
    if(num>threshold):
        num=1
    else: 
        num=0
    return num
    
```

```python
step(0.1)
```

## sigmoid function

```python
import math as m

def sigmoid(step):
    sig=(1/(1+m.exp(-step)))
    return sig
```

```python
sigmoid(500)
```

## Error Function

```python
def Err(z,exp):
        e=z*(1-z)*(exp-z)
        return e   
```

## WeightUpdate Function

```python
def weightupdate(w,z,error):
    w = w +(error*z)
    return w
```

# Epoches

```python
print("x1     x2")
for epoches in range(20):
        print("EPOCH:{}".format(epoches+1))
        itration=0
        for x1 in range(2):
            for x2 in range(2):
                print('x1:{}   x2:{}'.format(x1,x2))
                
                y1=(w1*x1)+(w3*x2)
                y1=sigmoid(y1)
                print('y1:{}'.format(y1))
                y2=(w4*x2)+(w2*x1)
                y2=sigmoid(y2)
                print('y2:{}'.format(y2))
                
                z=(w5*y1)+(w6*y2)
                z=sigmoid(z)
                print('z:{}'.format(z))
                
                error=Err(z,exp[itration])
                print('ERROR:{}'.format(error))
                
                w1=weightupdate(w5,z,error)
                print("weightUpdate  :{}".format(w1))
                w2=weightupdate(w6,z,error)
                print("weightUpdate  :{}".format(w2))
                
                hidnErr1=(error*w5)*((1-z)*z)
                hidnErr2=(error*w6)*((1-z)*z)
                
                w1 = w1 + (hidnErr1 * x1)
                w2 = w2 + (hidnErr2 * x2)
                w3 = w3 + (hidnErr1 * x1)
                w4 = w4 + (hidnErr2 * x2)
                
                print("new W1={}".format(w1))
                print("new W2={}".format(w2))
                print("new W3={}".format(w3))
                print("new W4={}".format(w4))
            
                itration=itration+1
            
            
            
        
```
